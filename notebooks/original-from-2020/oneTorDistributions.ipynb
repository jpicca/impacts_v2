```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
tors = pd.read_csv('./1950-2018_actual_tornadoes.csv')
```

```python
tors0918 = tors[(tors['yr'] > 2008) & (tors['mag'] > -1)]
```

```python
tors0918
```

```python
sns.distplot(tors0918['mag'],kde=True,rug=False,bins=np.arange(0,7),kde_kws={'clip':(0,6),'gridsize':6,'bw':0.5})
```

```python
apr27 = tors[tors['date'] == '2011-04-27']
```

```python
sns.distplot(apr27['mag'],kde=True,rug=False,bins=np.arange(0,7),kde_kws={'clip':(0,6),'gridsize':6,'bw':0.5})
```
