## Code/Plans for Stats Creation Script

- File initially delivered includes:
    - Counts of hospitals/mobile homes/power plants/population impacted for each sim
    - The med/90/10 of each state and cwa for hospitals/mobile homes/power plants/pop impacted
    - The daily climo of med/90/10 of states and cwas for hosp/mob/pow/pop
    
    Json example:
        {'sims': [[10,30,200,3002],[20,30,50,3043] ... [30,505,405,2302]],  <-- each nested list are the counts from an individual sim 
        'states':[{'state':'Alabama',
                   'hosp':[[23,230,5],[10,15,30]], <-- 1st inner list is med, 90, 10 of model run and 2nd is the climo of med, 90, 10 (stored in a file somewhere and grabbed)
                   'mob': [...],[...]
                   'pow': [...],[...]
                   'pop': [...],[...]
                   }, 
                   ...
                   {'state:'Wyoming',
                   ....}
                   ],
        'cwas':[{'cwa':'ABR',
                ....
                },
                ...
                {'cwa':'VEF',
                ....}
                ]
        }
- Then have individual files for CWAs and states that include the individual sim data for that particular area
    - Example: tx_sims.csv = sim, hosp, mob, pop, pow

### Get counts from the sims on national level

```python
masterDict = {}
```

Read model file and format the dataframe so that it includes one row for each simulation

```python
f = pathlib.Path("..","visualization","dashboard","backend","output","201905201626_tsigtest.psv")
#f = '../visualization/dashboard/backend/output/201905201626.psv.gz'
#f = pathlib.Path('.','visualization','initialDashboard','includes','data','199708291630_pp.psv')
df = pd.read_csv(f, sep="|")
sims = df.groupby("sim")
fields = sims.sum().loc[:,('population','hospitals','mobilehomes','psubstations')]
# Fill missing sims with 0s
fill_fields = fields.reindex(list(range(1,10001)),fill_value=0)
```

Get the tornadoes associated with a particular quantile

```python
quants = fill_fields.quantile(q=[0,0.1,0.5,0.9,1],interpolation='nearest')
sim = fill_fields[fill_fields['population'] == quants.loc[0.5,'population']].index[0]

df[df['sim'] == sim].loc[:,['slon','slat','elon','elat']].values.tolist()
df
```

Get tornado counts by simulation

```python
# torsBySim = df.groupby("sim").count().loc[:,'wfos']
# merged = fill_fields.merge(torsBySim,how='left',on='sim')
# merged.fillna(value=0,inplace=True)
# merged.rename(columns={'wfos':'tors'},inplace=True)
# merged

def torCounter(df,filled_df):
    # Use arbitrary column to get count of entries for each sim (i.e., count of tornadoes)
    torsBySim = df.groupby("sim").count().loc[:,'wfos']
    
    # Merge this count with the original 
    merged = filled_df.merge(torsBySim,how='left',on='sim')
    merged.fillna(value=0,inplace=True)
    merged.rename(columns={'wfos':'tors'},inplace=True)
    return merged

merged = torCounter(df,fill_fields)
```

Add the model data to the master dictionary and pull nat climo data from the climo file

```python
masterDict['sims'] = merged.values.tolist()
# These values are placeholders until actual climo is calculated
# Climo values will be stored in a file, grabbed, placed into this dict each day/run
# Note that we don't produce climo data for the # of tornadoes
masterDict['natClimo'] = {'pop': {},
                         'pow': {},
                         'mob': {},
                         'hosp': {}}
```

Read climo data

```python
with open('../climoFiles/pop_climo_smAvg.json') as file:
    pop_data = json.load(file)
    
with open('../climoFiles/hosp_climo_smAvg.json') as file:
    hosp_data = json.load(file)
    
with open('../climoFiles/mob_climo_smAvg.json') as file:
    mob_data = json.load(file)
    
with open('../climoFiles/pow_climo_smAvg.json') as file:
    pow_data = json.load(file)
```

Function to get index of current day file so we can grab correct data

```python
aggregateMonths = [0, 31, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335];

def getDayIdx(file):
    dayStr = str(file).split('/')[-1].split('.')[0][4:8]
    
    month = int(dayStr[:2])
    day = int(dayStr[2:])

    return aggregateMonths[month-1]+day-1
```

```python
dayIdx = getDayIdx(f)

for key in pop_data['nat'].keys():
    #print(key)
    #print(pop_data['nat'][key][dayIdx])
    
    masterDict['natClimo']['pop'][key] = pop_data['nat'][key][dayIdx]
    masterDict['natClimo']['hosp'][key] = hosp_data['nat'][key][dayIdx]
    masterDict['natClimo']['mob'][key] = mob_data['nat'][key][dayIdx]
    masterDict['natClimo']['pow'][key] = pow_data['nat'][key][dayIdx]
```

### Get stats from states

```python
stBrokenOut = df.assign(category=df['states'].str.split(',')).explode('category').reset_index(drop=True)

# Remove row if state is NaN
stBrokenOut = stBrokenOut[stBrokenOut['category'].notna()]

# Grab a list of the unique states in the simulation
statesImpacted = stBrokenOut['category'].unique().tolist()
```

```python
state = 'NY'

ind_state = stBrokenOut[stBrokenOut['category'] == state]
print(f'State: {state}')
ind_state_group = ind_state.groupby("sim").sum().loc[:,['population','hospitals','mobilehomes','psubstations']]
ind_state_tot = ind_state_group.reindex(list(range(1,10001)),fill_value=0)
```

```python
ind_state_tot.describe(percentiles=[0.1,0.5,0.9]).loc[['min','10%','50%','90%','max']]
```

Helper dictionary

```python
parmHelper = {
    'population': pop_data,
    'hospitals': hosp_data,
    'mobilehomes': mob_data,
    'psubstations': pow_data
}
```

Get stats from states impacted by any sim

```python
# Get sims for an individual state
masterDict['states'] = []

for state in statesImpacted:
    
    stateDict = {
        'state': state,
        'population': [],
        'hospitals': [],
        'mobilehomes': [],
        'psubstations': [],
        'tors': []
    }
    ind_state = stBrokenOut[stBrokenOut['category'] == state]
    print(f'State: {state}')
    ind_state_group = ind_state.groupby("sim").sum().loc[:,['population','hospitals','mobilehomes','psubstations']]
    ind_state_tot = ind_state_group.reindex(list(range(1,10001)),fill_value=0)
    
    ind_state_tot = torCounter(ind_state,ind_state_tot)
    
    ind_state_tot.to_csv(f'./{state}_data.csv')

    for impact in ['hospitals','population','mobilehomes','psubstations','tors']:
        stats = ind_state_tot.describe(percentiles=[0.1,0.5,0.9]).loc[['min','10%','50%','90%','max'],[impact]][impact].values.tolist()
        
        stateDict[impact].append(stats)
        
        # Append climo (gathered from a climo file)
        ## **** Need function logic to get climo for this particular state ****
        if impact == 'tors':
            break

        stateDict[impact].append([parmHelper[impact]['states'][state]['min'][dayIdx],
                                 parmHelper[impact]['states'][state]['ten'][dayIdx],
                                 parmHelper[impact]['states'][state]['med'][dayIdx],
                                 parmHelper[impact]['states'][state]['ninety'][dayIdx],
                                 parmHelper[impact]['states'][state]['max'][dayIdx]])

    masterDict['states'].append(stateDict)
    
#stBrokenOut.groupby(['category','sim']).sum().groupby('category').mean()
    
```

```python
ind_state = stBrokenOut[stBrokenOut['category'] == 'OK']
ind_state_group = ind_state.groupby("sim").sum().loc[:,['population','hospitals','mobilehomes','psubstations']]
ind_state_tot = ind_state_group.reindex(list(range(1,10001)),fill_value=0)
ind_state_tot.to_csv('./OK_data.csv',index=False)
```

### Get stats from CWAs

```python
wfoBrokenOut = df.assign(category=df['wfos'].str.split(',')).explode('category').reset_index(drop=True)

# Remove row if cwa is NaN
wfoBrokenOut = wfoBrokenOut[wfoBrokenOut['category'].notna()]

# Grab a list of the unique states in the simulation
wfoImpacted = wfoBrokenOut['category'].unique().tolist()
```

```python
# Get sims for an individual state
masterDict['cwas'] = []

for wfo in wfoImpacted:
    cwaDict = {
        'cwa': wfo,
        'population': [],
        'hospitals': [],
        'mobilehomes': [],
        'psubstations': [],
        'tors': []
    }
    ind_cwa = wfoBrokenOut[wfoBrokenOut['category'] == wfo]
    print(f'CWA: {wfo}')
    ind_cwa_group = ind_cwa.groupby("sim").sum().loc[:,['population','hospitals','mobilehomes','psubstations']]
    ind_cwa_tot = ind_cwa_group.reindex(list(range(1,10001)),fill_value=0)
    
    ind_cwa_tot = torCounter(ind_cwa,ind_cwa_tot)
    
    ind_cwa_tot.to_csv(f'./{wfo}_data.csv',index=False)

    for impact in ['hospitals','population','mobilehomes','psubstations','tors']:
        stats = ind_cwa_tot.describe(percentiles=[0.1,0.5,0.9]).loc[['min','10%','50%','90%','max'],[impact]][impact].values.tolist()
        
        cwaDict[impact].append(stats)
        
        if impact == 'tors':
            break
        
        # Append climo (gathered from a climo file)
        ## **** Need function logic to get climo for this particular state ****
        ## In meantime, placeholder values are present
        cwaDict[impact].append([parmHelper[impact]['cwas'][wfo]['min'][dayIdx],
                                 parmHelper[impact]['cwas'][wfo]['ten'][dayIdx],
                                 parmHelper[impact]['cwas'][wfo]['med'][dayIdx],
                                 parmHelper[impact]['cwas'][wfo]['ninety'][dayIdx],
                                 parmHelper[impact]['cwas'][wfo]['max'][dayIdx]])

    masterDict['cwas'].append(cwaDict)
```
