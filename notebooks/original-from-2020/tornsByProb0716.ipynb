```python
import pandas as pd
from sklearn.ensemble import IsolatedForest 
```

```python
#pd.read_csv('../output/201905201626.psv.gz',compression='gzip',sep='|')
def prepDF():
    df = pd.read_csv('../impacts-data/pas-input-data/torns-by-prob.csv')
    df['date'] = pd.to_datetime(df['date'],format='%Y%m%d')
    
    levels = ['00','02','05','10','15','30','45','60']
    for level in levels:
        df[f"d{level}"] = df[f"t{level}"]/df[f"a{level}"]
        
    return df
```

```python
df = prepDF()
```
